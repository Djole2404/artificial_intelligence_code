2. Are there any null values in the Grc_df? If yes, then in which columns and how many?
Finally, handle these null values using any strategy shown during the labs.

```python
Grc_df["Item_Weight"].replace(np.nan, Grc_df["Item_Weight"].mean(), inplace=True) #Replace all the Nan values with the mean of "Item_Weight"

Grc_df['Outlet_Size'] = Grc_df['Outlet_Size'].astype(str) # Change the type for "Outlet size from object to string"
Grc_df['Outlet_Size'] = Grc_df['Outlet_Size'].replace({'nan':'Medium'}) # Replace all the Nan values with the the most frequent value

Grc_df.isna().sum()
```

5. Drop the columns having index values of 0, 6 and create a new DataFrame Grc_new_df.

```python
Grc_new_df = Grc_df.drop(Grc_df.columns[[0, 6]], axis = 1) #making the new df without column 0 and 6
Grc_new_df.head(3) #print to check 
```

6. Using different Supermarket type listed in the column Outlet_Type create two different
DataFrames from Grc_new_df. Name these DataFrames as SupType_1 and SupType_2.

```python
#create the two new sub dataFrames for supermarket type 1 and supermarket type 2
SupType_1 = Grc_new_df[Grc_new_df['Outlet_Type'] == 'Supermarket Type1'] 
SupType_2 = Grc_new_df[Grc_new_df['Outlet_Type'] == 'Supermarket Type2']
```

7. Using Seaborn (“ggplot style”) create a (2,1) subplot of a box plot showing 5-point
summary of the column Item_MRP for SupType_1 and SupType_2. Which Outlet Type has
a higher median MRP? Are there any outliers?

```python
plt.style.use('ggplot')

fig, ax = plt.subplots(2, 1, figsize=(10,10)) #create a figure and axes with 2 rows and 1 column
sb.boxplot(x=SupType_1['Item_MRP'], ax=ax[0]) #create a boxplot of Item_MRP for SupType_1
ax[0].set_title('Boxplot of Item_MRP for Supermarket Type1')

sb.boxplot(x=SupType_2['Item_MRP'], ax=ax[1]) #create a boxplot of Item_MRP for SupType_2
ax[1].set_title('Boxplot of Item_MRP for Supermarket Type2')

#Display the plot
plt.show()
```

Supermarket Type 1 has a higher median MRP which is seen by the line in the box. There are no outliers judging by the graphs.

8. Concatenate the DataFrames SupType_1 and SupType_2 to create a new DataFrame
Grc_Concat_df and sort it in ascending order based on the column Item_Outlet_Sales.
What is the Outlet_Location_type of the store with the lowest sales?

```python
Grc_Concat_df = pd.concat([SupType_1, SupType_2]).sort_values("Item_Outlet_Sales") 
Grc_Concat_df.head() #printing to check if it looks right

```

```python
lowest_sales = Grc_Concat_df["Outlet_Location_Type"].head(1).values[0] 
print(f"The tier of the market with lowest sales is {lowest_sales}")
```

9. Using columns Outlet_Size and Outlet_Location_Type to create a multiple index for
Grc_Concat_df.

```python
Grc_Concat_df.set_index(['Outlet_Size', 'Outlet_Location_Type'], inplace=True, drop=False)
Grc_Concat_df.head(10)
```

10. Cut the column Item_Weight of Grc_Concat_df into 10 buckets and compute each bucket's
mean, minimum, maximum, and count. 

```python
#Cutting the Item_Weight column into 10 buckets
Grc_Concat_df['Buckets'] = pd.qcut(Grc_Concat_df['Item_Weight'], q=10)

#Group by the Weight_Bucket
stats = Grc_Concat_df.groupby('Buckets')['Item_Weight'].agg(
    mean='mean',
    min='min',
    max='max',
    count='count'
)

# Display the result
print(stats)
```
